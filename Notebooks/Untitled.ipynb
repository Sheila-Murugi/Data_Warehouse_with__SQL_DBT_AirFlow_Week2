```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv('20181101_d1_0800_0830.csv',low_memory=False)
display (df)
```

```python
# Specify the columns for each DataFrame
VehicleInfo = ['track_id', ' type',' traveled_d',' avg_speed']  # Columns for DataFrame 1
TrajectoryData = ['track_id',' lat', ' lon',' speed',' lon_acc',' lat_acc',' time']  # Columns for DataFrame 2

# Split the original DataFrame into two DataFrames
VehicleInfo = df.loc[:, VehicleInfo]
TrajectoryData = df.loc[:, TrajectoryData]

# Print the first few rows of each DataFrame
print("Vehicle Information:")
display (VehicleInfo)
print("\nTrajectory Data:")
display (TrajectoryData)
```
